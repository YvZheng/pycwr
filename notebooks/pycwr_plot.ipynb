# Using pycwr for ploting data

## read radar data

```python
%matplotlib inline
from pycwr.io import read_auto
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from pycwr.draw.RadarPlot import Graph, GraphMap
```

```python
PRD = read_auto(r"E:\RadarBaseData\WSR98D\Z9898\Z_RADR_I_Z9898_20190828181529_O_DOR_SAD_CAP_FMT.bin.bz2")
```

## Plotting PPI with map

```python
ax = plt.axes(projection=ccrs.PlateCarree())
graph = GraphMap(PRD, ccrs.PlateCarree()) ##
graph.plot_ppi_map(ax, 0, "dBZ", cmap="pyart_NWSRef") ## 0, dBZ
ax.set_title("Using pycwr for ploting data with map", fontsize=16)
plt.show()
```

## Plotting PPI

```python
fig, ax = plt.subplots()
graph = Graph(PRD)
graph.plot_ppi(ax, 0, "dBZ", cmap="pyart_NWSRef") ## 0, dBZcmap
graph.add_rings(ax, [0, 50, 100, 150, 200, 250, 300])
ax.set_title("Using pycwr for ploting data", fontsize=16)
ax.set_xlabel("Distance From Radar In East (km)", fontsize=14)
ax.set_ylabel("Distance From Radar In North (km)", fontsize=14)
plt.show()
```

```python
fig, ax = plt.subplots()
graph = Graph(PRD)
graph.plot_ppi(ax, 0, "ZDR") ## 0, ZDR
graph.add_rings(ax, [0, 50, 100, 150, 200, 250, 300])
ax.set_title("Using pycwr for ploting data", fontsize=16)
ax.set_xlabel("Distance From Radar In East (km)", fontsize=14)
ax.set_ylabel("Distance From Radar In North (km)", fontsize=14)
plt.show()
```
